```python
#with threading

from threading import Thread

def print_java(): #independent function
    for i in range(100):
        print("java")
        
def print_python(): #independent function
    for i in range(100):
        print("python")
        
def print_groovy(): #independent function
    for i in range(100):
        print("groovy")
        
        
thread1 = Thread(target = print_java)
thread2 = Thread(target = print_python)
thread3 = Thread(target = print_groovy)


thread1.start()
thread2.start()
thread3.start()

```

```python

def decorator_function(func):
    def inner_function():
        print("Lines to be executed before func1")
        func()
        print("Lines to be executed after func1")
    return inner_function

def func1(): #main function to be executed
    print("main functionality")


func1 = decorator_function(func1)

func1()
```

```python
def decorator_function(func):
    def inner_function():
        print("Lines to be executed before func1")
        func()
        print("Lines to be executed after func1")
    return inner_function

@decorator_function
def func1():
    print("main functionality")

func1()
```

```python
#example 2

# importing libraries
import time
import math

def calculate_time(func):

    def inner1(*args, **kwargs):

        begin = time.time()
        print("The time before factorial calculation is ", begin)

        func(*args, **kwargs)
        end = time.time()
        print("The time after factorial calculation is ", end)
        print("Total time taken in : ", func.__name__, end - begin)

    return inner1
@calculate_time
def factorial(num):
    time.sleep(10)
    print(math.factorial(num))
    
#factorial = calculate_time(factorial)

factorial(10)
```

## Dynamic args in python *args **kwargs

```python
def func1(a,b):
    return a + b


print(func1(3,4))
```

```python
def func1(a,b,c=10):
    return a + b + c


print(func1(3,4))
print(func1(3,4,5))
```

```python
def func1(*args):
    sum = 0
    for i in args:
        sum = sum + i
    return sum

print(func1(3,4))
print(func1(3,4,5))
print(func1(3,4,2,3,4,5,6,7,8,10))
print(func1(3))
```

```python
def func1(**kwargs):
    sum = 0
    for i in kwargs.values():
        sum = sum + i
    return sum

print(func1(val1 = 3, val2 = 4))
print(func1(val1 = 3,val2 = 4,val3 = 5))
print(func1(val1 =3))
```
